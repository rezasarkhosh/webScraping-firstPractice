```python
from bs4 import BeautifulSoup
import requests
```

```python
html_text = requests.get('https://www.apple.com/shop/buy-mac').text
print(html_text)
```

```python
soup = BeautifulSoup(html_text, 'lxml')
products = soup.find_all('div', class_='rf-hcard-content-title')
product_p= soup.find_all('span', class_='nowrap')
for name, value in zip(products, product_p[:-2]):
    product_price = value.text.strip()
    product_name = name.text.strip()
    print(f'{product_name} : {product_price}')
```

```python
import csv

products = soup.find_all('div', class_='rf-hcard-content-title')

with open('results.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Product Name', 'Product Price'])  # نوشتن سرستون‌ها

    for name, value in zip(products, product_p[:-2]):
        product_price = value.text.strip()
        product_name = name.text.strip()
        writer.writerow([product_name, product_price])  # نوشتن ردیف

print("نتایج با موفقیت در فایل CSV ذخیره شد.")


```
